🌟 Exercise 7: Transformation Exercise
For each of the following unstructured data sources, propose a method to convert it into structured data. Explain your reasoning.

A series of blog posts about travel experiences.
we can structure blogs data as following: country / location, date of visit, days of traveling, cities list , activities, attraction places, restaurants lists, etc.. store them into a table, rating(1-10)
Reason: 
Blog posts are free text, but they often contain repeated patterns: place names, experiences, opinions, rating, etc..


Audio recordings of customer service calls.
convert audio to text, extract keywords, categorize questions or topics and structure them into : Issues, Resolution, duration, satisfaction score
Audio itself is not directly analyzable, but once converted to text, we can extra patterns and information

Handwritten notes from a brainstorming session.
scanning the notes and extra texts through image extraction or (Optical Character Recognition) , once we get text, we can extract data again like: name of the person, keywords, theme, ideas
Brainstorming notes are: messy fragmented unordered . But they usually contain: ideas, tasks, themes

A video tutorial on cooking.
Extract audio → speech-to-text (recipe narration), Extract key frames (images of steps), Combine into structured recipe steps, then we can organize data into: steps, ingredients, duration, cooking tools needed, etc..
The video contains multi data types, audio, images, timing, By combining these, we can convert the video into a step-by-step recipe dataset.
