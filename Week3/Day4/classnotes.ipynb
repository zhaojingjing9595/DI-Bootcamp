```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y, color='green', linestyle='dashed', linewidth=2, marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
values = [20, 35, 30, 15]

plt.bar(categories, values)
plt.title('Fruit Sales Data')
plt.xlabel('Fruit')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np
from scipy import integrate

# Define a simple function
def integrand(x):
    return x**2

# Compute the integral of the function
result, _ = integrate.quad(integrand, 0, 1)
print("Integral of x^2 from 0 to 1:", result)
```
