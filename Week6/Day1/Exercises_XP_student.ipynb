1. Traditional ML and deep learning differ mainly in feature engineering:

Traditional ML (e.g. linear regression, SVM, random forests): a human manually selects and engineers the relevant features from raw data before feeding them to the model. The model then learns a mapping from those hand-crafted features to the output. It typically needs less data and less compute, and works well on structured/tabular data.

Deep learning (neural networks with multiple layers): the model automatically learns hierarchical feature representations directly from raw data (pixels, audio waveforms, raw text) — no manual feature engineering needed. Earlier layers learn simple patterns (edges, word co-occurrences), later layers combine them into complex, abstract features. This makes it especially powerful for unstructured data (images, audio, text), but it generally requires much more data, more compute (often GPUs), and is less interpretable ("black box").

2. How do artificial neural networks mimic the human brain?  
- Neurons as units: The brain has biological neurons connected by synapses; ANNs have artificial "neurons" (nodes) connected by weighted links. Each node receives inputs, combines them, and produces an output — analogous to a biological neuron receiving electrical/chemical signals and firing based on a threshold.

- Activation/firing: Biological neurons fire an electrical signal once input stimulation crosses a threshold. Artificial neurons apply an activation function (ReLU, sigmoid, etc.) to their weighted input sum to decide what signal to pass forward — a simplified analog of firing.

- Weighted connections = synaptic strength: In the brain, synapses strengthen or weaken based on experience (Hebbian learning — "neurons that fire together, wire together"). In ANNs, the weights on connections are numeric values that get adjusted during training (via backpropagation) to strengthen or weaken the influence of one node on another.

- Layers = hierarchical processing: The brain processes information in stages (e.g., visual cortex: simple edges → shapes → objects). Deep networks similarly stack layers, with early layers capturing simple patterns and deeper layers combining them into more abstract representations.

- Learning from experience: Both systems improve at a task through exposure to data/experience rather than explicit programming — the brain via neuroplasticity, ANNs via gradient-based training on examples.


3. Why does deep learning perform better on large datasets compared to traditional machine learning?  
- Model capacity scales with data: Deep networks have many layers and millions of parameters, giving them huge representational capacity. With small datasets this capacity causes overfitting, but with large datasets there's enough signal to fill that capacity with genuinely useful patterns rather than noise. Traditional ML models (e.g. linear models, shallow trees) have limited capacity — they plateau in performance because they simply can't represent more complex patterns, no matter how much data you feed them.

- Automatic feature learning needs data to work: Since deep learning learns its own features from raw data (rather than relying on hand-crafted ones), it needs many examples to discover which patterns are actually meaningful. With little data, there isn't enough signal to learn good features, so hand-crafted features (as used in traditional ML) tend to win. With lots of data, learned features can capture subtler, more complex, and more task-specific patterns than what humans manually design.

- Better generalization at scale: More data means the model sees more variation, so it learns more robust, generalizable representations instead of memorizing quirks of a small sample. This directly reduces overfitting risk for high-capacity models.

- Hierarchical representations compound with scale: Each additional layer builds on the abstractions of the previous one. With enough data to train all those layers properly, the network can capture very complex, hierarchical relationships (e.g. edges → shapes → objects) that shallow models can't express regardless of dataset size.

- Diminishing/plateauing returns for traditional ML: Traditional ML algorithms often saturate — after a certain amount of data, adding more barely improves performance, because the model's structure caps what it can learn. Deep learning curves tend to keep improving with more data (given enough compute), which is why it dominates on large-scale problems like image and language tasks.

5. What is feature engineering, and why is it not needed in deep learning? 
- Feature engineering is the process of using domain knowledge to manually create, select, or transform raw data into input variables ("features") that make it easier for a machine learning model to find patterns and make accurate predictions.

Deep neural networks perform automatic representation learning: instead of a human deciding what features matter, the network learns its own useful features directly from raw data as part of training.


6. What role do hidden layers play in a deep learning model? 

Their core role is progressive, hierarchical feature transformation. Each hidden layer takes the output of the previous layer, applies a weighted linear combination followed by a non-linear activation function (ReLU, sigmoid, tanh, etc.), and passes the result forward.


7. In an artificial neural network, what is the function of an activation function?

An activation function is applied to the weighted sum of inputs at each neuron, deciding what output that neuron passes forward to the next layer. Its main roles: 
- Introduces non-linearity
    Without an activation function, a neuron's output is just a linear combination of its inputs (weights · inputs + bias). Stacking multiple linear layers with no non-linearity in between is mathematically equivalent to a single linear layer — no matter how many layers you add, the network could only learn linear relationships. Activation functions break this linearity, letting the network approximate complex, non-linear real-world patterns (images, language, etc.).

- Decides whether/how strongly a neuron "fires"
Loosely mimicking biological neurons firing above a threshold, the activation function transforms the raw weighted sum into a bounded or shaped output — e.g., squashing it into a range, zeroing out negative values, etc.

- Enables gradient-based learning
Most activation functions are differentiable, which is required for backpropagation — the network needs to compute gradients of the loss with respect to weights, and that flows back through each activation function.

\[ \text{Weighted Sum} = 0.6\cdot \text{Temperature} + 0.4\cdot \text{Rain} + \text{Bias} \]
Apply a step activation:

```python
# To-Do: implement weighted_sum then use it to compute the two cases
def weighted_sum(temperature_f: float, rain01: int) -> float:
    """Compute s = 0.6*Temperature + 0.4*Rain + Bias."""
    # To-Do: replace 'pass' with the formula using W_TEMP, W_RAIN, BIAS
    s = W_TEMP * temperature_f + W_RAIN * rain01 + BIAS
    return s

# To-Do: compute decisions for the two cases and print them
case1 = (70, 0)  # Temperature=70, Rain=0
case2 = (50, 1)  # Temperature=50, Rain=1

s1 = weighted_sum(*case1)
s2 = weighted_sum(*case2)
y1 = step_activation(s1)
y2 = step_activation(s2)

print({"case1": {"sum": s1, "decision": y1}, "case2": {"sum": s2, "decision": y2}})
```

## 🌟 Exercise 3: Building a Simple Neural Network with TensorFlow or Keras

**As stated in the instructions**  
Build a classifier for MNIST with:
- One input layer
- One hidden layer with 128 neurons and ReLU activation
- One output layer with 10 neurons and softmax activation

Steps:
1. Import TensorFlow and needed modules from `tensorflow.keras`
2. Load the MNIST dataset
3. Normalize data by dividing by 255.0
4. One-hot encode the labels
5. Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
6. Compile with optimizer `adam`, loss `categorical_crossentropy`, metrics `accuracy`
7. Train and evaluate on the test set

**Guidance**  
You will execute prefilled boilerplate for imports and dataset loading. You will complete the one-hot encoding and the model specification, compile, train, and evaluate.

**Learning point**  
Normalization stabilizes optimization. One-hot encoding matches softmax outputs to target vectors. ReLU accelerates training by avoiding saturation, while softmax maps logits to a probability simplex.

![image.png](https://github.com/user-attachments/assets/61737b80-69cb-41a1-aea5-76092cd2a373)


```python
# PREFILLED: just execute
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist

print("TensorFlow:", tf.__version__)

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Flatten will be added in the model; keep images as 28x28 for now
print("Train shape:", x_train.shape, "Test shape:", x_test.shape)
```

```python
# To-Do: one-hot encode labels y_train and y_test using to_categorical
# Hints: use to_categorical
y_train_oh = to_categorical(y_train, num_classes=10)
y_test_oh  = to_categorical(y_test, num_classes=10)

print("Labels one-hot:", y_train_oh.shape, y_test_oh.shape)
```

```python
# To-Do:  Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
model = models.Sequential([
    layers.Input(shape=(28, 28)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax'),
])

model.summary()
```

```python
# To-Do: compile the model with adam, categorical_crossentropy, and accuracy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# To-Do: train then evaluate
history = model.fit(x_train, y_train_oh, epochs=5, batch_size=32, validation_split=0.1, verbose=1)
test_loss, test_acc = model.evaluate(x_test, y_test_oh)
print({"test_loss": float(test_loss), "test_acc": float(test_acc)})
```

## 🌟 Exercise 5 (optional): Implementing Forward and Backward Propagation in Python

**As stated in the enoncé**  
You will code a simple neural network that performs forward propagation and backpropagation for a regression problem predicting exam scores from study hours.

**Guidance**  
Run the prefilled code. Then modify the learning rate or initial weights and observe changes in the loss and updates.
Provide a short written explanation.

**Learning point**  
Gradient descent updates parameters in the direction of steepest loss reduction.
The learning rate controls step size. Too small slows learning. Too large risks divergence.


```python
# PREFILLED: just execute
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```
