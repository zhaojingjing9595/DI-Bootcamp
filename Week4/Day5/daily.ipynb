The basics of machine learning is a subset of AI where algorithms learn patterns from data in order to help specific purposes like making prediction or decisions, finding patterns or outliers, find relationships between data, or create new output, etc...

Examples of ML across industries:
1. online social media, shopping websites or business track your past behaviors to find patterns and predict your preferences, need or future behavior
2. map or navigation system: based on datasets to predict travel duration , traffic jams on different hours of the day, etc..
3. Biology or chemical field: protein folding issue, using existing proven dataset to predict other protein folding 

Supervised ML: given input and output ( labeled data), let machine figure out the way to predict on new input
for example: the machine will be trained based on a given sets of cat/dog images with labels on each image, so that it will find patterns.

Unsupervised ML: only given input, let the machine to learn and figure out the differences between data, clustering or dimensionality reduction, without an output guidance
for example: the machine will be trained with a given sets of images without any label or output information, and it needs to find the relations / objects in the images

reinforcement learning: given an interactive environment, a set of possible actions for machine to act on, by giving feedback like reward or penalty on each action so that the machine learns how to find the optimal way to reach the goal.
for example: we can train the machine with a chess game , by giving the environment and sets of basic rules and give points based on each action, so it can learn on how to get a higher score through trail and error.

Process of developing a machine learning model: 
1. Define an issue or statement or a purpose
2. feature selection: choose dataset that fits the purpose, and selecting relevant features reduces noise, prevents overfitting, and improves model performance. Data understanding, cleaning, and analysis techniques are needed
3. model selection: choose proper model based on your goal and data. (e.g., linear regression for continuous output, decision trees for interpretability, neural networks for complex patterns).
4. model evaluation: use training data to train the model and test the results, do iterations and adjustment for the model, and evaluation uses metrics appropriate to the problem: accuracy, precision/recall, F1-score for classification; RMSE/MAE for regression.
