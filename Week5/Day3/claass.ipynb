```python
import pandas as pd
df = pd.read_csv('Housing.csv')

df.head()
```

```python
df = pd.get_dummies(df, drop_first=True, dtype=float)
df.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(df['price'])
plt.show()
```

```python
df = df[(df['price']<=9000000) & (df['area'] <=10500) & (df['bedrooms'] < 6)]
df.info()

df.head()
```
